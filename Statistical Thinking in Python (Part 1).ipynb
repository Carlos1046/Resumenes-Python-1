```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()

datos = pd.read_csv("https://assets.datacamp.com/production/course_1549/datasets/michelson_speed_of_light.csv")

datos.head(1)
```

```python
datos.iloc[:,3].head()
```

Creando un Histograma para conocer el comportamiento de la temperatura 

```python
#Es buena practica calcular los grupos adecuados para visualizar mejor el histograma.
n_data = len(datos)
n_bins = int(np.sqrt(n_data))

_ = plt.hist(datos.iloc[:,3], bins = n_bins)
_ = plt.xlabel('temperature F')
_ = plt.ylabel('Count')
plt.show()
```

Creacion de un grafico por parcelas y  con los datos representados en enjambre de abejas

```python
datos['date'].value_counts()
```

```python
ax = sns.swarmplot(x = 'date', y = 'temperature (F)', data = datos)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
```

Funciones de distribucion acumulativa empirica

```python
def ecdf(data): 
    n = len(data)
    
    x = np.sort(data)
    y = np.arange(1, n+1)/n
    return x,y
```

```python
x_temp, y_temp = ecdf(datos.iloc[:,3])

percentiles = np.array([2.5,25,50,75,97.5])
ptiles_tem = np.percentile(datos.iloc[:,3],[2.5,25,50,75,97.5])

np.mean(datos.iloc[:,3])
```

```python
_ = plt.plot(x_temp, y_temp, marker = '.', linestyle = 'none')
_ = plt.xlabel('Temperature F')
_ = plt.ylabel('Acumulado')

_ = plt.plot(ptiles_tem, percentiles/100, marker = 'D', color = 'red', linestyle = 'none')

plt.show()
```

```python
bx = sns.boxplot(x = 'date', y = 'Cor', data = datos)
bx.set_xticklabels(bx.get_xticklabels(),rotation=30)
plt.show()
```

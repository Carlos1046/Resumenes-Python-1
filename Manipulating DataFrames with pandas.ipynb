```python
import pandas as pd
import numpy as np
election = pd.read_csv('https://assets.datacamp.com/production/repositories/502/datasets/502f4eedaf44ad1c94b3595c7691746f282e0b0a/pennsylvania2012_turnout.csv',index_col= 'county')
```

```python
election.head()
```

```python
#.iloc (Especificamos numericamente la fila y columna) y .loc (Especificamos con nombres la fila y columna)
election.iloc[4, 4] == election.loc['Bedford', 'winner']
```

```python
#Selecion de Columnas
election[['winner','voters']].head()
```

```python
#Seleccion de columnas y elementos del 0:5
election['winner'][0:5]
```

```python
#Todas las filas y seleccion entre 2 Columnas
election.loc[:, 'Obama':'voters'].head()
```

```python
#Seleccion entre filtas y todas las columnas
election.loc['Armstrong':'Bedford',:]
```

```python
#Seleccion entre filas y columnas
election.loc['Armstrong':'Bedford','Obama':'voters']
```

```python
#Seleccion entre filas 2:5 y de la columna 1 en adelante
election.iloc[2:5,1:]
```

```python
#Seleccion entre filas y ciertas columnas
election.loc['Armstrong':'Berks',['Obama','voters']]
```

```python
#Ciertas filas y columnas
election.iloc[[1,4,5],[2,4]]
```

```python
#Diferencia de objeto al seleccionar, Series (Estructura de datos para una sola columna)
#dataframe (Estructura de datos tabular)
#Por lo tanto las series son las columnas y dataframes diccionarios
type(election['Obama'])
```

```python
type(election[['Obama']])
```

```python
#Creando listas de valores a tomar en cuenta
rows = ['Philadelphia', 'Centre', 'Fulton']
cols = ['winner', 'Obama', 'Romney']

election.loc[rows, cols]
```

### Filtros

Necesitamos crear condiciones logicas dentro de nuestros datos.

```python
(election.Obama > 40).head()
```

```python
#Toma solo los valores True de la condicion logica generada.
election[election.Obama > 40].head()
```

```python
#Podemos generar nuestras reglas en una lista y luego pasasela al dataframe
Filtro = election.Obama > 40
election[Filtro].head()
```

```python
#Multiples condiciones (Y se define con &), (O se define como |)
election[(election.Obama > 40) & (election.winner == 'Obama')].head()
```

```python
too_close = election['margin'] < 1

#Asigna NaN, donde se cumpla la condicion
election.loc[too_close, 'winner'] = np.nan
election.head(8)
```

```python
too_close = election['Romney'] < 45

#Asigna 0, donde se cumpla la condicion
election.loc[too_close, 'winner'] = 0
election.head(8)
```

### Deteccion Valores Faltantes

```python
election.notnull().all() #Con que uno sea False regresa False
```

```python
election.notnull().any() #any() Si almenos un valor es True regresa todo como True
```

```python
election.isnull().any() #Regresa True donde tegamos un nulo
```

```python
election.dropna(how = 'any').shape #Si almenos en alguna fila existe algun NaN, la elimina
```

```python
election.dropna(how = 'all').shape #Toda la fila debe tener NaN
```

```python
election.dropna(thresh = 66, axis='columns').shape #Elimina la columna, que contenga 66 valores no nulos
```

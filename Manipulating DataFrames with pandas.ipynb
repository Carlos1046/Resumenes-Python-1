### Filtros

Necesitamos crear condiciones logicas dentro de nuestros datos.

```python
(election.Obama > 40).head()
```

```python
#Toma solo los valores True de la condicion logica generada.
election[election.Obama > 40].head()
```

```python
#Podemos generar nuestras reglas en una lista y luego pasasela al dataframe
Filtro = election.Obama > 40
election[Filtro].head()
```

```python
#Multiples condiciones (Y se define con &), (O se define como |)
election[(election.Obama > 40) & (election.winner == 'Obama')].head()
```

```python
too_close = election['margin'] < 1

#Asigna NaN, donde se cumpla la condicion
election.loc[too_close, 'winner'] = np.nan
election.head(8)
```

```python
too_close = election['Romney'] < 45

#Asigna 0, donde se cumpla la condicion
election.loc[too_close, 'winner'] = 0
election.head(8)
```

### Deteccion Valores Faltantes

```python
election.notnull().all() #Con que uno sea False regresa False
```

```python
election.notnull().any() #any() Si almenos un valor es True regresa todo como True
```

```python
election.isnull().any() #Regresa True donde tengamos un nulo
```

```python
election.dropna(how = 'any').shape #Si almenos en alguna fila existe algun NaN, la elimina
```

```python
election.dropna(how = 'all').shape #Toda la fila debe tener NaN
```

```python
election.dropna(thresh = 66, axis='columns').shape #Elimina la columna, que contenga 66 valores no nulos
```

### Transformaciones dataFrame

.apply() Permite aplicar una funcion arbitraria a cada elemento.

```python
def Operacion(n):
    return np.log(n) * 100
```

```python
election[['total','Romney','voters']].apply(Operacion).head()
```

.map() transforma los valores de acuerdo a una busqueda en los diccionarios.

```python
red_vs_blue = {'Obama':'blue', 'Romney':'red'}
election['color'] = election['winner'].map(red_vs_blue)
election.head()
```

Usando funciones Vectorizadas, Cuando el rendimiento es primodial evitemos el uso de .apply() y .map()
Numpy y SciPy vienen con una variedad de funciones vectorizadas.

```python
from scipy.stats import zscore

#En lugar de usar .apply(), tenemos funciones en scipy que ayudan. ejemplo zscore (Estad)
turnout_zscore = zscore(election['turnout'])
print(type(turnout_zscore))
election['turnout_zscore'] = turnout_zscore
election.head()
```

Los indices son objetos inmutables. Significa que si se desea cambiar o modificar el indice de un Data Frame, es necesario cambiar todos los valores.

No apoyaremos en las listas de comprension para esto.

```python
new_idx = [names.upper() for names in election.index]
print(new_idx[0:5])
election.index = new_idx
election.head()
```

```python
#Colocaremos un nombre a nuestros indices y columnas
election.index.name = 'County'
election.columns.name = 'Resultados'
election.head()
```

Tambien se puede construir los indices desde cero, y luego juntarlos con el marco de datos.
Se debe tener cuidado para evitar errores de matcheo.

```python
new_idx = range(len(election))
election_copy = election.copy()
election_copy.index = new_idx
election_copy.head()
```

### Reorganizacion de los datos

Pivoteando una sola variable.

```python
gapminder = pd.read_csv('https://assets.datacamp.com/production/repositories/502/datasets/09378cc53faec573bcb802dce03b01318108a880/gapminder_tidy.csv')
gapminder.head()
```

Nos gustaria tener un resumen de la pertenencia de los paises por region, al pivotear una variable no podemos tener datos duplicados.

Por lo que solo trabajaremos con un año 1964

```python
gapminder = gapminder.set_index(['Country','Year']) 
gapminder_copy = gapminder.loc[(slice(None),1964),:] #Filtramos solo los valores del año 1964 para no tener datos duplicados
gapminder_copy = gapminder_copy.reset_index() #.pivot no permitira trabajar con los index
```

```python
gapminder_copy.pivot(index = 'region', columns='Country', values = 'gdp')
```

```python
#Si no especificamos la columna a pivotear todas las de mas columnas seran pivoteadas
gapminder_copy.pivot(index = 'region', columns = 'Country')
```

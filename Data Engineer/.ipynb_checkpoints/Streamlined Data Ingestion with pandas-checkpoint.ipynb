### Importing Data from Flat Files

Tips para facilitar la importación de datos por medio de un csv (tomar ciertas columnas, comenzar por una fila, customizar valores Nan, determinar el tipo de datos, errores tokenizing)

```python
import pandas as pd
import matplotlib as plt

archivo = "https://assets.datacamp.com/production/repositories/4412/datasets/61bb27bf939aac4344d4f446ce6da1d1bf534174/vt_tax_data_2016.csv"

data = pd.read_csv(archivo)

data.head()
```

```python
data.shape
```

```python
#La mayoría de nuestros datos son de tipo entero, ¿Y si quiero crear variables categorias o strings, desde la importación?
data.dtypes.head(10)
```

```python
counts = data.groupby("agi_stub").N1.sum()
counts.plot.bar()
```

```python
#Reduciendo el número de columnas que desamos ocupar
cols = ["zipcode", "agi_stub", "mars1", "MARS2", "NUMDEP"]

pd.read_csv(archivo, usecols=cols).groupby("agi_stub").sum()
```

```python
pd.read_csv(archivo, nrows = 100, skiprows = 100, header = None).head()
```

```python
# Para cambiar el tipo de datos creamos un diccionario para especificar por culumna el tipo de datos que deseamos
data_types = {"agi_stub": "category","zipcode": str}

data = pd.read_csv(archivo, dtype=data_types)

data.dtypes.head(10)
```

```python
#Podemos customizar los datos NA, obervemos que la columna zipcode ceros. Indiquemos que esos valores son faltantes Nan

null_values = {"zipcode":0}

data = pd.read_csv(archivo, 
                   na_values = null_values, usecols = cols)

print(data[data.zipcode.isna()])
```

### Importing Data From Excel Files

Tips para facilitar la lectura de un archivo de excell. (Hojas, concatención)

```python
archivo = 'https://assets.datacamp.com/production/repositories/4412/datasets/fdb113aa942a0e0ad5c155b2f04784886f0038c9/fcc-new-coder-survey.xlsx'

col_string = "AD, AB, AW:BA"
#a diferencia de un csv tenemos que cuidar el nombre de las hojas, nombres de las columnas, posición de lectura y las celdas de interés.
survey_response = pd.read_excel(archivo, skiprows = 2, usecols = col_string, sheet_name = 1)

job_prefs = survey_response.groupby("JobPref").JobPref.count()
job_prefs.plot.barh()

survey_response.head()
```

```python
#hojas de un archivo
all_survey_data = pd.read_excel(archivo,
                                sheet_name=None, skiprows = 2)

print(all_survey_data.keys())
```

```python
all_responses = pd.DataFrame()

# Contaremos el tamaño de los datos en cada hoja
for df in all_survey_data.values():
  # numero de filas por hoja
  print("Adding {} rows".format(df.shape[0]))
  # Append df to all_responses, assign result
  all_responses = all_responses.append(df)

#Grafica uniendo las 2 hojas
counts = all_responses.groupby("EmploymentStatus").EmploymentStatus.count()

counts.plot.barh()
```

Tipos de datos, con dtypes = {columna_i : tipo_dato}  ó parse_dates = [columna_fecha]
